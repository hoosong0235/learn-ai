```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/ARMY/learn-ai/machineLearning/Part 4 - Clustering/Section 24 - K-Means Clustering/Python"
dataset = pd.read_csv(path + "/Mall_Customers.csv")
X = dataset.iloc[:, 3:].values
```

```python
# Using the elbow method to find the optimal number of clusters

from sklearn.cluster import KMeans
wcss = list()
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Training the K-Means model on the dataset

kmeans = KMeans(n_clusters=5, init="k-means++", random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualising the clusters

cs = ["red", "blue", "green", "cyan", "magenta"]
for i in range(5):
  plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], c=cs[i], label="Cluster {}".format(i))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c="yellow", label="Centroids")
plt.title("Clusters of customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
