```python
# Importing the libraries

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# Preprocessing the Training set

from tensorflow.keras.preprocessing.image import ImageDataGenerator
path = "/content/drive/MyDrive/ARMY/learn-ai/deepLearning/Part 2 - Convolutional Neural Networks/dataset"
train_datagen = ImageDataGenerator(
  rescale=1./255,
  shear_range=0.2,
  zoom_range=0.2,
  horizontal_flip=True,
)
train_set = train_datagen.flow_from_directory(
  path + "/training_set",
  target_size=(64, 64),
  batch_size=32,
  class_mode='binary'
)
```

```python
from tensorflow.keras.utils import image_dataset_from_directory
path = "/content/drive/MyDrive/ARMY/learn-ai/deepLearning/Part 2 - Convolutional Neural Networks/dataset"
train_set = image_dataset_from_directory(
    path + "/training_set",
    image_size=(64, 64),
    batch_size=32,
    label_mode="binary"
)
```

```python
# Preprocessing the Test set

from tensorflow.keras.preprocessing.image import ImageDataGenerator
path = "/content/drive/MyDrive/ARMY/learn-ai/deepLearning/Part 2 - Convolutional Neural Networks/dataset"
test_datagen = ImageDataGenerator(
  rescale=1./255,
  shear_range=0.2,
  zoom_range=0.2,
  horizontal_flip=True,
)
test_set = train_datagen.flow_from_directory(
  path + "/test_set",
  target_size=(64, 64),
  batch_size=32,
  class_mode='binary'
)
```

```python
from tensorflow.keras.utils import image_dataset_from_directory
path = "/content/drive/MyDrive/ARMY/learn-ai/deepLearning/Part 2 - Convolutional Neural Networks/dataset"
train_set = image_dataset_from_directory(
    path + "/test_set",
    image_size=(64, 64),
    batch_size=32,
    label_mode="binary"
)
```

```python
# Initializing the CNN

cnn = tf.keras.models.Sequential()
```

```python
# Step 1 - Convolution

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=(64, 64, 3)))
```

```python
# Step 2 - Pooling

cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=2))
```

```python
# Adding a second convolutional layer

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu"))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=2))
```

```python
# Step 3 - Flattening

cnn.add(tf.keras.layers.Flatten())
```

```python
# Step 4 - Full Connection

cnn.add(tf.keras.layers.Dense(units=128, activation="relu"))
```

```python
# Step 5 - Output Layer

cnn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Compiling the CNN

cnn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Training the CNN on the Training set and evaluating it on the Test set

cnn.fit(x=train_set, validation_data=test_set, epochs=25)
```

```python
# Making a single prediction

from tensorflow.keras.preprocessing import image
path = "/content/drive/MyDrive/ARMY/learn-ai/deepLearning/Part 2 - Convolutional Neural Networks/dataset"
test_image = image.load_img(path + "/single_prediction/cat_or_dog_1.jpg", target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
cnn.predict(test_image)
```
